```python
import numpy as np
np.random.seed(1)
# function to select random values
# picks 1000 random ints btw 0 to 50
x = np.random.randint(0,50,100)
x
```

```python
#positive corelation with same noise
y = x + np.random.normal(0,10,100)
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
#negative corelation with same noise
y = 100 - x + np.random.normal(0,10,100)
np.corrcoef(x,y)
```

```python
x = np.random.normal(0,10,100)
y = np.random.normal(0,10,100)
```

```python
# Correlation matrix
import pandas as pd
```

```python
df = pd.DataFrame({'a' : np.random.randint(0,50,1000)})
df['b']=df['a'] + np.random.randint(0,10,1000)
df['c']= 100 - df['a'] + np.random.randint(0,10,1000)
df['d']= np.random.randint(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
